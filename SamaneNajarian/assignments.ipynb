```python
# Assignment3

for i in range(2, 1001):
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            break
    else:
        print(f'{i} is a prime number.')
```

```python
#Assignment5

numbers = []

while True:
    number = input('Enter an integer (or "Q" to Quit): ')
    if number.upper() == 'Q':
        break
    elif number.isdigit():
        numbers.append(number)
    else:
        print("Please enter a valid integer.")
        continue

total = 0
count = 0
for num in numbers:
    total += int(num)
    count += 1

if count > 0:
    mean = total / count
    print("Mean:", mean)
else:
    print("No numbers entered.")

print("="*60)

import statistics

numbers = []
while True:
    num = input('Enter an integer (or "Q" to Quit): ')
    if num.upper() == 'Q':
        break
    if num.isdigit():
        numbers.append(int(num))
    else:
        print("Please enter a valid integer.")

if numbers:
    meanValue = statistics.mean(numbers)
    print("Mean:", meanValue)
else:
    print("No numbers entered.")
```
